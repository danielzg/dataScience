```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (30, 16))

dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
#Para fixar uma área para o graficos
area = plt.figure()
```

```python
# A área irá possuir 2 linhas e 2 colunas e o g1 ocupará o 1º item da matriz
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor X Área')

g2.hist(dados['Valor'])
g2.set_title('Histograma do Valor')

dadosG3 = dados['Valor'].sample(100)
dadosG3.index = range(dadosG3.shape[0])
g3.plot(dadosG3)
g3.set_title('Amostra do Valor')

grupoG4 = dados.groupby('Tipo')['Valor']
label = grupoG4.mean().index
valores = grupoG4.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```
