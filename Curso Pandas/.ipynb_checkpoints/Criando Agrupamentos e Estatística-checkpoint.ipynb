```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguelResidencialTipoAgregado')
dados.head(12)
```

```python
dados['Valor'].mean()
```

```python
dados['Bairro'].unique()
```

```python
bairros = ['Copacabana', 'Jardim Botânico', 'Centro', 'Higienópolis',
      'Cachambi', 'Barra da Tijuca', 'Ramos', 'Grajaú',
     'Lins de Vasconcelos', 'Botafogo']

selecaoBairros = dados['Bairro'].isin(bairros)
dadosBairro = dados[selecaoBairros]
dadosBairro.head(10)
```

```python
dados['Valor'].mean()
```

```python
dadosBairro['Valor'].mean()
```

```python
dados.shape[0]
```

```python
dadosBairro.shape[0]
```

```python
dadosBairro['Bairro'].value_counts()
```

```python
grupoBairros = dadosBairro.groupby('Bairro')
```

```python
type(grupoBairros)
```

```python
grupoBairros.groups
```

```python
for bairro, dados in grupoBairros:
    #print(bairro)
    #print(dados)
    #print(type(dados))
    print('Bairro:{} -> Media de Valores {} '.format(bairro, dados['Valor'].mean()))
```

```python
#Jeito mais facil de fazer o mesmo que a celula anterior
grupoBairros[['Valor', 'Condominio']].mean().round(2)
```

```python
grupoBairros['Valor'].describe().round(2)
```

```python
grupoBairros['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Mínimo', 'max' : 'Máximo', 'sum' : 'Soma'})
```

# MATPLOTLIB

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupoBairros['Valor'].std().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Desvio Padrão')
```

```python
fig = grupoBairros['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_xlabel('Valor Medio do Aluguel por Bairro', {'fontsize' : 22})
```
