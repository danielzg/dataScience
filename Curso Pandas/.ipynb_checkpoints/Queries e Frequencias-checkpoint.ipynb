# Seleções e Frequências

```python
import pandas as pd

dados = pd.read_csv('dados/aluguelResidencial.csv')
dados.head(13)
```

```python
#Selecionar somente os imóveis classificados com 'Tipo' = 'Apartamento'
selecaoApartamento = dados['Tipo'] == 'Apartamento'
n1 = dados[selecaoApartamento].shape[0]
n1
```

```python
#Selecionar somente os imóveis classificados com 'Tipo' = 'Casa', 'Casa de Condominio' ou 'Casa de Vila'
selecaoCasa = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecaoCasa].shape[0]
n2
```

```python
#Selecionar os imoveis com áreas entre 60 e 100m^2, incluindo os limites
selecaoArea = (dados['Area'] >= 60) & (dados['Area'] <= 100)
n3 = dados[selecaoArea].shape[0]
n3
```

```python
#Selecionar os imoveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2000
selecaoQuartosAluguel = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecaoQuartosAluguel].shape[0]
n4
```

```python
#Exercicios
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
alunos
```

```python
#Crie um DataFrame somente com os alunos aprovados
selecaoAprovados = alunos['Aprovado'] == True
dfAprovados = alunos[selecaoAprovados]
dfAprovados
```

```python
#Crie um DataFrame somente com as alunas aprovadas.
selecaoAlunasAprovadas = (alunos['Aprovado'] == True) & (alunos['Sexo'] == 'F')
dfAlundasAprovadas = alunos[selecaoAlunasAprovadas]
dfAlundasAprovadas
```

```python
#Crie apenas uma visualização dos alunos com idade entre 10 e 20 anos ou com idade maior ou igual a 40 anos.
selecaoIdade = ((alunos['Idade'] >= 10) & (alunos['Idade'] <= 20)) | (alunos['Idade'] >= 40)
dfIdade = alunos[selecaoIdade]
dfIdade
```
